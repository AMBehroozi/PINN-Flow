```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings
from google.colab import drive
# open train data from google drive
drive.mount('/content/drive')
```

```python
# import the data
dataset= pd.read_csv('/content/drive/MyDrive/minst/data.csv')
dataset['Date']=pd.to_datetime(dataset['Date'], format='%m/%d/%Y')
```

```python
# on our data, impute the missing values using the interpolation techniques and specifically, the lineare method
dataset['Fixed_Discarge']= dataset['Discarge'].interpolate(option='spline')
```

```python
import matplotlib.pyplot as plt
plt.close("all")
plt.figure();
plt.plot(dataset.Date, dataset.Discarge, color='blue', label="Discarge")
plt.plot(dataset.Date, dataset.Fixed_Discarge, color='red', label="Fixed Discarge")

# Giving title to the graph
plt.title('Historical River Discarge')
 
# rotating the x-axis tick labels at 30degree
# towards right
plt.xticks(rotation=30, ha='right')
plt.legend(loc="upper right")
# Giving x and y label to the graph
plt.xlabel('Date')
plt.ylabel('Discarge')
```
